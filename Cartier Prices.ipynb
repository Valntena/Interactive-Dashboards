```python
# App layout
app = JupyterDash(__name__)
app.layout = html.Div(children=[
    html.H1(children='Cartier Prices Dashboard'),
    dcc.Dropdown(id='categorie-dropdown',
                 options=[{'label': i, 'value': i}
                          for i in df['categorie'].unique()],
                 value='rings'),
    dcc.Graph(id='price-graph')
])
    
    
    # Set up the callback function
@app.callback(
    Output(component_id='price-graph', component_property='figure'),
    Input(component_id='categorie-dropdown', component_property='value')
)


def update_graph(selected_categorie):
    filtered_df = df[df['categorie'] == selected_categorie]
    line_fig = px.scatter(filtered_df,
                       x='price', y='title',
                       color='tags',
                       title=f'cartier Prices in {selected_categorie}')
    return line_fig



# Run local server
if __name__ == '__main__':
    app.run_server(port=8051)
```
